## Dropping irrelevant columns

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

## Renaming the columns

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

## Dropping duplicate rows

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

## Dropping missing or null values

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

## Detecting Outliers
Using 1.5 x IQR method and checking with box plots

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

As we can see from the above box plots that the data is normally distributed
So, we can safely use 1.5 x IQR method to remove outliers

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

## Visualizing data using plots
